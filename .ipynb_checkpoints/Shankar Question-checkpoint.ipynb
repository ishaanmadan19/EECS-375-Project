```python


# df = df.drop("subject_age", axis =1)
df = pd.read_csv("MACHINE LEARNING/trr__officer_combined__new.csv")
# df = df.fillna(0)
# df = df.drop("subject_birth_year", axis = 1)
# df = df.drop("birth_year", axis = 1)
# df = df.drop("direction", axis = 1)
cols = df.columns
g = df.columns.to_series().groupby(df.dtypes).groups   
for key in g:
#     print(key)
    if key == 'int64':
        for col in g[key]:
#             print(key, col)
            df[col] = df[col].fillna(value = 0)
    if key == 'float64':
        for col in g[key]:
#             print(key, col)
            df[col] = df[col].fillna(value = 0.0)
    if key == 'object':
        for col in g[key]:
#             print(key, col)
            df[col] = df[col].fillna('0')

# df = df.drop("indoor_or_outdoor", axis = 1)
# df = df.drop("lighting_condition", axis = 1)
# df = df.drop("weather_condition", axis = 1)
df = df.apply(LabelEncoder().fit_transform)
df.to_csv("Encoded_data.csv")
```

```python
df = df.drop(["trr_id", "resistance_level"], axis = 1)
X = df.drop("action_category", axis = 1)
Y = df["action_category"]
clf_gini = DecisionTreeClassifier(criterion = "gini", max_depth=10, min_samples_leaf=5)
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.3)
# acc = []
# for i in range(0,100):
#     pred_gini = clf_gini.predict(X_test)
#     acc.append(accuracy_score(y_test,pred_gini)*100)

scores_DTG = cross_val_score(clf_gini, X, Y, cv=10)
```
